PROBLEM #1

Which NYC schools have the best math results?

The best math results are at least 80% of the *maximum possible score of 800* for math.
Save your results in a pandas DataFrame called best_math_schools, including "school_name" and "average_math" columns, sorted by "average_math" in descending order.

```python
# Reduce the schools dataframe to only include schools that have an average math score better than 80%
av_math_better_than_80_percent = schools[schools['average_math'] > (800 * 0.80)]
# 10 schools have an average math score better than 80%
# now create a best_math_schools dataframe and reduce the rows to school_name and average_math only. We can also sort by the avergae math column and ensure the data is displayed in descending order
best_math_schools = av_math_better_than_80_percent[['school_name', 'average_math']].sort_values(by='average_math', ascending=False)
print(best_math_schools)
```
